```python
import numpy as np 
from sympy import * 
def operations(A,rowchanged,rawchange,value):
    A = A.astype(float)
    A[rowchanged-1,:] = A[rawchange-1,:]*value+A[rowchanged-1,:]
    print(A)
    return A
def simpleoperation(A,row,value):
    A = A.astype(float)
    A[row-1,:] = A[row-1,:]/value
    print(A)
    return A
```
